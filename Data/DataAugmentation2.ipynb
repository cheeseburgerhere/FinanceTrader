```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
import os

def get_csvs(dir:str):
    dirList=os.listdir(dir)
    endResult=[]
    for dire in dirList:
        if(os.path.join(dir,dire).endswith(".csv")):
            endResult.append(dir+"/"+dire)
    return endResult 
```

```python
allDirs=get_csvs("./StandardizedData")
```

```python
indexes=get_csvs("./StandardizedBIST")
```
